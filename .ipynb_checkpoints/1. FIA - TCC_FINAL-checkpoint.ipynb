## 1. Dataset Import
The original file is a csv using "|" as a delimitor of columns

````python
def import_file(p_file):
    file_path = "files/" + p_file
    
    with open(file_path, "r") as file:
        file_treated = [x.strip() for x in file.readlines()]
        file_read = [x.split("|") for x in file_treated]
    
    return file_read
````

````python
def create_df_upon_csv_file(p_content):

    #create the dataframe using the list
    df = pd.DataFrame(p_content)

    # grab the first row and define as header
    header = df.iloc[0]

    # consider the data, but the first row as data
    df = df[1:]
    df.columns = header

    return df
````

### 2.4 Check the missings
All columns are analysed, counting the rows and null rows.

````python
def count_missings(df):
    list_missings = []
    for col in df.columns.tolist():
        column = col
        count = df[col].count()
        count_missing = df[col].isnull().sum()
        if count_missing != 0:
            list_missings.append(column)
        print(f'{column} : {count}  =>  Nulls: {count_missing}')
    return(list_missings)
````

### 4.2 Create the age group

````python
def define_age_group(row):
    if row["AGE"] < 25:
        return "1"
    elif row['AGE'] >= 25 and row['AGE'] <= 35:
        return '2'
    elif row['AGE'] > 35 and row['AGE'] <= 45:
        return '3'
    elif row['AGE'] > 45:
        return '4'
    else:
        return '5'
    
def define_age_group_description(row):
    if row["AGE"] < 25:
        return "1. Less than 25"
    elif row['AGE'] >= 25 and row['AGE'] <= 35:
        return '2. Between 25 and 35'
    elif row['AGE'] > 35 and row['AGE'] <= 45:
        return '3. Between 36 and 45'
    elif row['AGE'] > 45:
        return '4. More than 45'
    else:
        return '5. No Information'
````

````python
df_dataset_treated["YEAR_ORDER_DATE"] = pd.to_datetime(df_dataset_treated["ORDER_DATE"]).dt.year
df_dataset_treated["AGE"] = df_dataset_treated["YEAR_ORDER_DATE"] - df_dataset_treated["BIRTHDATE_YEAR"]
df_dataset_treated["AGE_GROUP"] = df_dataset_treated.apply(define_age_group, axis=1)
df_dataset_treated["AGE_GROUP_DESCRIPTION"] = df_dataset_treated.apply(define_age_group_description, axis=1)
display(df_dataset_treated.head(2))
````

The median is different from median, which tell us that there are outliers, which we can see that are before 10% and after 90%. Thus, we will consider the median for the ages among these classifications.

````python
def adjust_age_group(row, median):
    if row["AGE"] < 18:
        return median
    elif row['AGE'] >= 18 and row['AGE'] < 60:
        return row["AGE"]
    elif row['AGE'] >= 60:
        return median
    else:
        return median
````

### 4.8 Transformantion - Null Treatment 

The replacement of null data is done considering the following rules:
* String -> "No information"
* Numeric -> Average

````python
def replace_missings(df):
    for col in df:
        if col in ["ORDER_ID", "ORDER_ID_ITEM", "ORDER_STATUS", "ACCOUNT_ID", "USER_ID", "SKU", "STYLE_COLOR_CODE", "PRODUCT", "COLOR", "SIZE", "DEPARTMENT_PRODUCT", "CATEGORY_PRODUCT", "TYPE_PRODUCT", "AGE_GROUP_DESCRIPTION", "SEASON", "TYPE_USER"]:
            df[col] = df[col].replace(np.NaN, "NO INFORMATION") 
        elif col in ["ORDER_DATE", "GRADE", "QTY_PURCHASES_PROD", "QTY_N_PURCHASES_PROD", "PRODUCT_CODE", "RATING"]:
            df[col] = df[col]
        else:
            df[col] = df[col].fillna(pd.to_numeric(df_dataset[col], errors='coerce').mean()).astype(float)
    return df
````

### 6.2.1 Model Based -> K-Nearest Neighbours (k-NN Basic)
It is available in Surprise as KNNBasic and is a basic collaborative filtering algorithm.
The prediction can be done using one of the following formulas, depending on the approach: user-based or item-based.

<img src="images/knn_basic.png">

Doing a tunning to check the best parameters:

* __name__ contains the similarity metric to use. Options are cosine, msd, pearson, or pearson_baseline. The default is msd.
* __user_based__ is a boolean that tells whether the approach will be user-based or item-based. The default is True, which means the user-based approach will be used.
* __min_support__ is the minimum number of common items needed between users to consider them for similarity. For the item-based approach, this corresponds to the minimum number of common users for two items.

Reference:   
https://surprise.readthedocs.io/en/stable/getting_started.html#tune-algorithm-parameters-with-gridsearchcv

````python
#Reference: https://surprise.readthedocs.io/en/stable/getting_started.html#train-test-split-and-the-fit-method
from surprise import Reader
from surprise import Dataset
from surprise import KNNBasic
from surprise import accuracy
from surprise.model_selection import train_test_split

reader = Reader(rating_scale=(1, 4))

# Loads Pandas dataframe
data_dataset = Dataset.load_from_df(df_dataset_modeling[["users", "items", "ratings"]], reader)
#display(data_dataset)

# sample random trainset and testset
# test set is made of 25% of the ratings.
trainset, testset = train_test_split(data_dataset, test_size=.3)

sim_options = {
    "name": ["cosine"],
    "min_support": [5],
    "user_based": [True],
}

param_grid = {"sim_options": sim_options}

#print(trainset)

# We'll use the famous KNNBasic algorithm.
algo = KNNBasic(sim_options=param_grid)

# Train the algorithm on the trainset, and predict ratings for the testset
algo.fit(trainset)

# Prediction using test
predictions = algo.test(testset)

# Then compute RMSE
accuracy.rmse(predictions)
````

So, for the approach:  

__Similarity metric:__ Cosine  
__Approach:__ User-based  
__Minimum number of common items needed for similarity:__ 5  

The RMSE is 0.77, which means that the model is might miss the prediction between -0.77 and 0.77. For instance, for a rating that is 2, the prediction can be 1.37 until 2.77

### 6.3.1 Model-based -> Matrix Factorization-based algorithms Singular Value Decomposition (SVD)
We can use Surprise to check which values for the following factors work best:

* __n_epochs__ is the number of iterations of SGD, which is basically an iterative method used in Statistics to minimize a function.
* __lr_all__ is the learning rate for all parameters, which is a parameter that decides how much the parameters are adjusted in each iteration.
* __reg_all__ is the regularization term for all parameters, which is a penalty term added to prevent overfitting.

Reference:  
https://surprise.readthedocs.io/en/stable/matrix_factorization.html#surprise.prediction_algorithms.matrix_factorization.SVD

````python
#Reference: https://surprise.readthedocs.io/en/stable/getting_started.html#train-test-split-and-the-fit-method
from surprise import SVD
from surprise import accuracy
from surprise.model_selection import train_test_split

# sample random trainset and testset
# test set is made of 30% of the ratings.
trainset, testset = train_test_split(data_dataset, test_size=.3)

# We'll use the SVD algorithm.
algo = SVD(n_epochs=15, lr_all=0.005, reg_all=0.4)

# Train the algorithm on the trainset, and predict ratings for the testset
algo.fit(trainset)
predictions = algo.test(testset)

# Then compute RMSE
accuracy.rmse(predictions)
````

### Check the results of predictions:
Referencee:  
https://towardsdatascience.com/building-and-testing-recommender-systems-with-surprise-step-by-step-d4ba702ef80b  
https://nbviewer.jupyter.org/github/NicolasHug/Surprise/blob/master/examples/notebooks/KNNBasic_analysis.ipynb  
https://surprise.readthedocs.io/en/stable/FAQ.html?highlight=SVD#how-to-get-the-top-n-recommendations-for-each-user

````python
from collections import defaultdict

def get_top_n(predictions, n=10):
    '''Return the top-N recommendation for each user from a set of predictions.

    Args:
        predictions(list of Prediction objects): The list of predictions, as
            returned by the test method of an algorithm.
        n(int): The number of recommendation to output for each user. Default
            is 10.

    Returns:
    A dict where keys are user (raw) ids and values are lists of tuples:
        [(raw item id, rating estimation), ...] of size n.
    '''

    # First map the predictions to each user.
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))

    # Then sort the predictions for each user and retrieve the k highest ones.
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]

    return top_n
````

````python
#Define the top number
top_n = get_top_n(predictions, n=2)

# Print the recommended items for each user
for uid, user_ratings in top_n.items():
    print(uid, [iid for (iid, _) in user_ratings])
````
